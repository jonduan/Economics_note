```python
%pylab inline
%config InlineBackend.figure_format = 'svg'

from fieldplot import fieldplot

def vectors():   # actual vector field
    global a,b,c,d
    def f(x,y): return a*x + b*y
    def g(x,y): return c*x + d*y
    fieldplot(f,g,-5,5,-5,5,nx=20)
    
def directions():  # direction field
    global a,b,c,d
    def f(x,y): return a*x + b*y
    def g(x,y): return c*x + d*y
    def fu(x,y): return f(x,y)/sqrt(f(x,y)**2+g(x,y)**2)
    def gu(x,y): return g(x,y)/sqrt(f(x,y)**2+g(x,y)**2)
    fieldplot(fu,gu,-5,5,-5,5,nx=20)
```

```python
a,b =  0,1
c,d = -1,0
vectors()
```
